```python
import cv2
import os
from ultralytics import YOLO
import shutil
from google.colab import files


model = YOLO('yolov8n.pt')
def detect_objects(img):
    results = model(img)
    detected_objects = []

    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            confidence = box.conf[0]
            class_id = int(box.cls[0])
            detected_objects.append((class_id, confidence, (x1, y1, x2, y2)))

    return detected_objects


video_path = "/content/drive/MyDrive/complete/Y2meta.app-Relaxing highway traffic-(720p).mp4"
cap = cv2.VideoCapture(video_path)
output_folder = "detected_objects"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
frame_number = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frame_number += 1
    detected_objects = detect_objects(frame)
    for idx, (class_id, confidence, box) in enumerate(detected_objects):
        x1, y1, x2, y2 = box
        crop_img = frame[y1:y2, x1:x2]
        output_path = os.path.join(output_folder, f"frame_{frame_number}_object_{idx}.jpg")
        cv2.imwrite(output_path, crop_img)
cap.release()
cv2.destroyAllWindows()
shutil.make_archive('detected_objects', 'zip', output_folder)
files.download('detected_objects.zip')
```

```python
import cv2
import os
from ultralytics import YOLO

# Load YOLOv8
model = YOLO('yolov8n.pt')  # or 'yolov8s.pt', 'yolov8m.pt', 'yolov8l.pt', 'yolov8x.pt'

# Function to detect objects in a frame
def detect_objects(img):
    results = model(img)
    detected_objects = []

    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            confidence = box.conf[0]
            class_id = int(box.cls[0])
            detected_objects.append((class_id, confidence, (x1, y1, x2, y2)))

    return detected_objects

# Read video and process frames
video_path = "/content/drive/MyDrive/complete/Y2meta.app-Relaxing highway traffic-(720p).mp4"
cap = cv2.VideoCapture(video_path)

output_folder = "detected_objects"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

frame_number = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_number += 1
    detected_objects = detect_objects(frame)

    for idx, (class_id, confidence, box) in enumerate(detected_objects):
        x1, y1, x2, y2 = box
        crop_img = frame[y1:y2, x1:x2]
        output_path = os.path.join(output_folder, f"frame_{frame_number}_object_{idx}.jpg")
        cv2.imwrite(output_path, crop_img)

cap.release()
cv2.destroyAllWindows()
```
